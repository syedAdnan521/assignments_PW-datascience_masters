```python
# Ans : 3 error defined in arithmatic class 
def print_exception_hierarchy(filter_class,level=0):
    print("  " * level + filter_class.__name__)
    for subclass in filter_class.__subclasses__():
        print_exception_hierarchy(subclass, level + 1)

if __name__ == "__main__":
    filtered_class = ArithmeticError
    print("\nFiltered Class:")
    print_exception_hierarchy(filtered_class)
```
